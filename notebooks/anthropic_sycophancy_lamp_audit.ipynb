# Public LLM Dataset LAMP Audit: Anthropic Sycophancy

Real public LLM-evaluation prompts from `Anthropic/model-written-evals` are used to audit prompt-only sycophancy-target monitors against answer-key leakage.

## Executive Result

The prompt-only TF-IDF monitor passes with AUC 0.987. The answer-key leakage monitor reaches AUC 1.000, but fails temporal isolation, forbidden-feature screening, and oracle proximity. This gives LAMP a real public LLM-data benchmark, not only a synthetic alignment toy.

| Model                                        | Type                      | AUC   | Audit Status | Key Failures                          |
| -------------------------------------------- | ------------------------- | ----- | ------------ | ------------------------------------- |
| TF-IDF prompt-only sycophancy-target monitor | valid prompt-only monitor | 0.987 | PASS         | none                                  |
| Interpretable user-view cue monitor          | valid cue monitor         | 0.815 | PASS         | none                                  |
| Visible length shortcut                      | visible shortcut          | 0.465 | FAIL         | null/destroyed                        |
| Answer-key leakage monitor                   | oracle leakage            | 1.000 | FAIL         | temporal, forbidden, oracle proximity |
| TF-IDF + answer-key leakage 1%               | low-dose oracle leakage   | 0.988 | FAIL         | temporal, forbidden, oracle proximity |
| TF-IDF + answer-key leakage 5%               | low-dose oracle leakage   | 0.992 | FAIL         | temporal, forbidden, oracle proximity |
| TF-IDF + answer-key leakage 10%              | low-dose oracle leakage   | 0.996 | FAIL         | temporal, forbidden, oracle proximity |
